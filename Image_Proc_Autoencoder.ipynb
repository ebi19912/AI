# Task
Create and train an autoencoder model using "/content/grayscale_images" as input and "/content/resized_images" as output.

## Load and preprocess data

### Subtask:
Load the grayscale input images and the resized color output images and preprocess them for the autoencoder model.


**Reasoning**:
Load the grayscale input images and the resized color output images and preprocess them for the autoencoder model.



```python
import numpy as np
import os
import tensorflow as tf
from PIL import Image

grayscale_dir = 'grayscale_images'
resized_dir = 'resized_images'

grayscale_files = sorted(os.listdir(grayscale_dir))
resized_files = sorted(os.listdir(resized_dir))

# Ensure filenames match
matching_files = [f for f in grayscale_files if f in resized_files]

input_images = []
output_images = []

for filename in matching_files:
    grayscale_path = os.path.join(grayscale_dir, filename)
    resized_path = os.path.join(resized_dir, filename)

    try:
        img_gray = Image.open(grayscale_path)
        img_resized = Image.open(resized_path)

        input_images.append(np.array(img_gray))
        output_images.append(np.array(img_resized))

        img_gray.close()
        img_resized.close()
    except Exception as e:
        print(f"Error processing {filename}: {e}")

input_images = np.array(input_images)
output_images = np.array(output_images)

# Preprocess images: scale to [0, 1]
input_images = input_images.astype('float32') / 255.0
output_images = output_images.astype('float32') / 255.0

# Reshape grayscale images to have a single channel
input_images = np.expand_dims(input_images, axis=-1)

print(f"Shape of input images: {input_images.shape}")
print(f"Shape of output images: {output_images.shape}")
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Model

# Define the input shape (grayscale image: height, width, 1 channel)
input_shape = (128, 128, 1)

# Encoder
input_img = Input(shape=input_shape)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

# Decoder
x = Conv2D(128, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x) # Output channels match the color image (3 channels)

# Autoencoder Model
autoencoder = Model(input_img, decoded)

# Print the model summary
autoencoder.summary()
```

```python
autoencoder.compile(optimizer='adam', loss='mse')
autoencoder.summary()
```

```python
history = autoencoder.fit(input_images, output_images,
                          epochs=10,
                          batch_size=32,
                          shuffle=True,
                          validation_split=0.2)
```

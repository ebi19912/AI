# Task
I have this Excel, I want to draw a chart that shows what the f1 score is for each model
My file has several sheets and each sheet is named after the model that was trained and the evaluation criteria are in each sheet for each label
Or if possible draw a chart like this for any other evaluation criteria

Here is all the data you need:
"Models compare.xlsx"

## Data loading

### Subtask:
Load the data from the "Models compare.xlsx" file into pandas DataFrames.


**Reasoning**:
I need to load the data from the Excel file into a dictionary of pandas DataFrames, where each key is a sheet name (model name) and the value is the corresponding DataFrame.



```python
import pandas as pd

try:
    all_models_data = pd.read_excel("Models compare.xlsx", sheet_name=None)
    print(f"Successfully loaded {len(all_models_data)} sheets.")
    # Display the sheet names
    print("Sheet names (model names):", list(all_models_data.keys()))
    # Example: Display the first few rows of the first sheet
    first_sheet_name = list(all_models_data.keys())[0]
    display(all_models_data[first_sheet_name].head())
except FileNotFoundError:
    print("Error: 'Models compare.xlsx' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data wrangling

### Subtask:
Extract and organize the evaluation metrics from each model's data.


**Reasoning**:
Extract the f1-scores (or a suitable alternative metric) from each model's DataFrame and store them in a nested dictionary, then convert this to a Pandas DataFrame.



```python
evaluation_metrics = {}
for model_name, df in all_models_data.items():
    f1_column = None
    for col in ['f1-score', 'Weighted F1-Score', 'f1']:
        if col in df.columns:
            f1_column = col
            break
    if f1_column:
        model_metrics = {}
        for index, row in df.iterrows():
            label = row['Label']
            f1_score = row[f1_column]
            model_metrics[label] = f1_score
        evaluation_metrics[model_name] = model_metrics
    else:
        # Handle the case where no suitable f1 column is found
        model_metrics = {}
        for index, row in df.iterrows():
            label = row['Label']
            model_metrics[label] = 0  # Assign 0 if no f1 column is found
        evaluation_metrics[model_name] = model_metrics
# Convert the dictionary to a DataFrame
evaluation_df = pd.DataFrame.from_dict(evaluation_metrics, orient='index')
display(evaluation_df.head())
```

## Data visualization

### Subtask:
Create a bar chart visualizing the f1-scores for each model across different labels.


**Reasoning**:
Create the bar chart visualizing the f1-scores for each model across different labels, and save it as "model_comparison_chart.png".



```python
import matplotlib.pyplot as plt

evaluation_df.plot(kind='bar', figsize=(15, 8))
plt.title('F1-Scores for Different Models across Labels')
plt.xlabel('Models')
plt.ylabel('F1-Score')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)
plt.tight_layout()
plt.savefig('model_comparison_chart.png', bbox_inches='tight')
plt.show()
```

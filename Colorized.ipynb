```python
# prompt: فایلهای jpg رو از داخل آدرس /content/dataset بردار و بزار توی پوشه images اگر اسم فایلی یکی بود تغییر بده تا تمام فایلها موجود باشه
# فقط فایلها رو میخوام و مخیوام تمام فایلها در یک پوشه باشه

import os
import shutil

# Create the 'images' directory if it doesn't exist
if not os.path.exists('/content/images'):
    os.makedirs('/content/images')

# Iterate through all files and subdirectories in '/content/dataset'
for root, _, files in os.walk('/content/dataset'):
    for file in files:
        if file.lower().endswith(('.jpg', '.jpeg')):  # Check for JPG files (case-insensitive)
            source_path = os.path.join(root, file)
            destination_path = os.path.join('/content/images', file)

            # Handle file name collisions
            i = 1
            while os.path.exists(destination_path):
                name, ext = os.path.splitext(file)
                destination_path = os.path.join('/content/images', f"{name}_{i}{ext}")
                i += 1

            shutil.copy2(source_path, destination_path) # copy2 preserves metadata
```

```python
# prompt: بیا و عکس های داخل پوشه images  رو تغییر اسم بده و به ترتیب همه رو شماره گذاری کن

import os

image_dir = '/content/images'

# Check if the directory exists
if not os.path.exists(image_dir):
    print(f"Error: Directory '{image_dir}' not found.")
else:
    # Get a list of image files in the directory
    image_files = [f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    image_files.sort() # Sort files for consistent numbering

    # Rename the files sequentially
    for i, filename in enumerate(image_files):
        name, ext = os.path.splitext(filename)
        new_filename = f"{i+1:04d}{ext}"  # Use zero-padding for consistent file names (e.g., 0001.jpg, 0002.jpg)
        source = os.path.join(image_dir, filename)
        destination = os.path.join(image_dir, new_filename)

        os.rename(source, destination)
        print(f"Renamed '{filename}' to '{new_filename}'")
```

```python
# prompt: عکس های داخل images رو به gray تبدیل کن در پوشه ای به همین اسم بزار

import cv2
import os

# Create the 'gray_images' directory if it doesn't exist
output_dir = '/content/gray_images'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Iterate through the images in the 'images' directory
for filename in os.listdir('/content/images'):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):  # Check for image files
        image_path = os.path.join('/content/images', filename)
        try:
            # Read the image using OpenCV
            img = cv2.imread(image_path)

            # Convert the image to grayscale
            gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            # Create the output file path
            output_path = os.path.join(output_dir, filename)

            # Save the grayscale image
            cv2.imwrite(output_path, gray_img)

        except Exception as e:
            print(f"Error processing image {filename}: {e}")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import os
from PIL import Image

# ابعاد تصاویر
img_width, img_height = 256, 256

# مسیر پوشه‌های تصاویر
rgb_dir = "/content/images"
gray_dir = "/content/gray_images"

# تابع بارگذاری تصاویر با حالت انتخابی
def load_images(directory, mode="rgb"):
    images = []
    for filename in os.listdir(directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            img = Image.open(os.path.join(directory, filename))
            img = img.resize((img_width, img_height))
            if mode == "gray":
                img = img.convert("L")  # اطمینان از بارگذاری به صورت سیاه و سفید
                img_array = np.array(img, dtype=np.float32) / 255.0
                # افزودن بعد کاناله (256,256) -> (256,256,1)
                img_array = np.expand_dims(img_array, axis=-1)
            else:
                img = img.convert("RGB")
                img_array = np.array(img, dtype=np.float32) / 255.0
            images.append(img_array)
    return np.array(images)

# بارگذاری تصاویر
rgb_images = load_images(rgb_dir, mode="rgb")
gray_images = load_images(gray_dir, mode="gray")

# تعریف معماری U-Net
def unet_colorization(input_shape=(img_width, img_height, 1)):
    inputs = keras.Input(shape=input_shape)

    # رمزگذار (Encoder)
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    c1 = layers.BatchNormalization()(c1)
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    c1 = layers.BatchNormalization()(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)

    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    c2 = layers.BatchNormalization()(c2)
    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    c2 = layers.BatchNormalization()(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)

    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    c3 = layers.BatchNormalization()(c3)
    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c3)
    c3 = layers.BatchNormalization()(c3)
    p3 = layers.MaxPooling2D((2, 2))(c3)

    c4 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(p3)
    c4 = layers.BatchNormalization()(c4)
    c4 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(c4)
    c4 = layers.BatchNormalization()(c4)
    p4 = layers.MaxPooling2D((2, 2))(c4)

    # Bottleneck
    c5 = layers.Conv2D(1024, (3, 3), activation='relu', padding='same')(p4)
    c5 = layers.BatchNormalization()(c5)
    c5 = layers.Conv2D(1024, (3, 3), activation='relu', padding='same')(c5)
    c5 = layers.BatchNormalization()(c5)

    # رمزگشا (Decoder)
    u6 = layers.UpSampling2D((2, 2))(c5)
    u6 = layers.Conv2D(512, (2, 2), activation='relu', padding='same')(u6)
    u6 = layers.concatenate([u6, c4])
    c6 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(u6)
    c6 = layers.BatchNormalization()(c6)
    c6 = layers.Conv2D(512, (3, 3), activation='relu', padding='same')(c6)
    c6 = layers.BatchNormalization()(c6)

    u7 = layers.UpSampling2D((2, 2))(c6)
    u7 = layers.Conv2D(256, (2, 2), activation='relu', padding='same')(u7)
    u7 = layers.concatenate([u7, c3])
    c7 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(u7)
    c7 = layers.BatchNormalization()(c7)
    c7 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c7)
    c7 = layers.BatchNormalization()(c7)

    u8 = layers.UpSampling2D((2, 2))(c7)
    u8 = layers.Conv2D(128, (2, 2), activation='relu', padding='same')(u8)
    u8 = layers.concatenate([u8, c2])
    c8 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u8)
    c8 = layers.BatchNormalization()(c8)
    c8 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c8)
    c8 = layers.BatchNormalization()(c8)

    u9 = layers.UpSampling2D((2, 2))(c8)
    u9 = layers.Conv2D(64, (2, 2), activation='relu', padding='same')(u9)
    u9 = layers.concatenate([u9, c1])
    c9 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u9)
    c9 = layers.BatchNormalization()(c9)
    c9 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c9)
    c9 = layers.BatchNormalization()(c9)

    # لایه خروجی: 3 کاناله (تصویر RGB)
    outputs = layers.Conv2D(3, (1, 1), activation='sigmoid')(c9)

    model = keras.Model(inputs, outputs)
    return model

# ساخت مدل U-Net
model = unet_colorization()
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.summary()

# استفاده از callbacks جهت بهبود روند آموزش
callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True),
    keras.callbacks.ModelCheckpoint("best_colorization_model.h5", save_best_only=True, monitor="val_loss")
]

# آموزش مدل (تقسیم بخشی از داده‌ها برای اعتبارسنجی)
history = model.fit(
    gray_images,
    rgb_images,
    validation_split=0.1,
    epochs=200,
    batch_size=16,
    callbacks=callbacks
)

# ذخیره مدل آموزش‌دیده
model.save("colorization_model.keras")
```
